```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FTS(T):    
    #return np.fft.fftshift( np.fft.fft(np.fft.fftshift(T)) )
    return np.fft.fftshift( np.fft.fft(np.fft.fftshift(T)) )
    #return np.fft.fft(T) 

def iFTS(T):
    return np.fft.ifftshift( np.fft.ifft(np.fft.ifftshift(T)) )    
    #return np.fft.ifft(T) 

def plh(dz, kz, r, T, lam, dx):
    """
    frensel impulse response propagation
    
    kz - фазовая маска
    T - поле, с которым производится операция """
    
    if (lam*dz/(T.shape[0]*dx**2)) < 0.3:
        f = iFTS( FTS(T)*(np.exp(1j*kz*dz)) )
        
    else:
        k = 2*np.pi/lam
        #fh = 1/(1j*lam*dz)*np.exp(1j*k*(r**2)/2/dz)
        fh = np.exp(1j*k*(r**2)/2/dz)
        fh = FTS(fh) #нормировка        
        f = iFTS( FTS(T)*fh )
    return f

def rphase(dx,L,k):
    kx = np.arange(-1/(2*dx), 1/(2*dx), 1/L)
    
    kz = 2*np.pi*np.sqrt((k/(2*np.pi))**2-kx**2)
    return kz

def Lens(x, k, focus):
    r2 = x**2
    return  k*r2/(2*focus)

def circ(r,a):
    return np.abs(r) < a
```

<b>Lens plane</b>

For lens sampling the relation for pixel size mast be fulfilled: $ dx \leq \frac{f}{2D}\lambda $

```python
lam = .85*10**-3
dx = 0.53*10**-3
l  = 256*2
L  = dx*l
k  = 2*np.pi/lam
x = np.arange(-L/2, L/2, dx)    

kz = rphase(dx,L,k)
r = np.abs(x)

sig = 1/np.sqrt(2)*40*10**-3
UL = np.exp(-r**2/2/sig**2).astype(complex)
fok = 200*10**-3
UL *= np.exp(-1j*Lens(x,k,fok))
plt.plot(np.abs(UL)*circ(r,.04))
# dx mmust be less (in mm)
print( '{:.2f}'.format(fok*lam/(2*(2*sig*2))*1000) )
```
